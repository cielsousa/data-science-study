```python
import pandas as pd
import numpy as np
import plotly.express as px

clients = pd.read_csv("C:\\Users\\Ciel\\Envs\\desafios\\Envs\\data-science-study\\desafio2\\clientes.csv", encoding='latin1')

clients.describe()
clients.info()

clients["Categoria Cartão"].unique()

# The column "Categoria Cartão" has one less row, we've to delete this
```

```python
# Deleting the 'nan' row
clients = clients.dropna(axis=0, how='any')

# Deleting the 'CLIENTNUM' column that isn't useful for our analyse
clients = clients.drop(axis=1, columns='CLIENTNUM')

clients.info()
```

```python
# Data Visualization

clients_fig = px.histogram(clients, x=clients['Categoria Cartão'], color=clients['Categoria'])
clients_fig.show()
```

```python
for i in clients.columns:
    clients_fig = px.histogram(clients, x=clients[i], color=clients['Categoria'], pattern_shape=clients['Sexo'], text_auto=True)
    clients_fig.show()
```
